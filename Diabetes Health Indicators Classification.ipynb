# **CDC Diabetes Health Indicators**
## **Overview**
Centers for Disease Control and Prevention (CDC) is an American government organization that serves purpose or protecting the public's health, whether diseases start at home or abroad, are chronic or acute, cureable or preventable. CDC puts science into action to help children stay health so they can grow and learn; to help families, businesses, and communities fight disease and stay strong. To accomplish their mission, CDC conducts critical science and provides health information that protects the people agains expensive and dangerous health threats and respond when these arise. 
<br>
<br>
The Behavioral Risk Factor Surveillance System (BRFSS) is a system or health-related telephone surveys that collect state data about US residents regarding their health-related risk behaviors, chronic health conditions, health-care access, and use of preventive services. The BRFSS surveys are supported by the CDC National Center for Chronic Disease Prevention and Health Promotion; other CDC centers; and federal agencies, such as the Health Resources and Services Administration, Administration on Aging, Department of Veterans Affairs, and Substance Abuse and Mental Health Services Administration.
<br>
<br>
One of the surveys the BRFSS done is regarding diabetes. About 38 million US adults have diabetes, and 1 in 5 of them don't know it, with higher prevalence among non-Hispanic Black (14.5%), Hispanic (12.5%), and Native American (14.7%) adults. It costs $327 billion annually and causes complications like heart disease, kidney failure, and blindness. Diverse risk factors make early detection challenging. Thus, the goal for this project is to build predictive models using BRFSS data to identify individuals at risk of diabetes enabling early intervention and supporting public health strategies.
<br>
<br>
The objectives of this project are:

1. **Given the health indicators, are they having diabetes?** - Develop ML models to identify predictive risk factor combinations.
2. **Which indicators best predict diabetes?** - Determine which feature in the dataset have the biggest predictive power.
3. **How to inform Public Health?** - Provide visualizations and tables to communicate risk factors and performance, guiding CDC prevention strategies.

## **Exploratory Data Analysis**

```python
# Correlation Matrix
def cramers_v(x, y):
    contingency_table = pd.crosstab(x, y)
    chi2, _, _, _ = chi2_contingency(contingency_table)
    n = contingency_table.sum().sum()
    phi2 = chi2 / n
    r, k = contingency_table.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))
```

```python
corr_matrix = pd.DataFrame(index = categorical_features, columns = categorical_features)
for col1 in categorical_features:
    for col2 in categorical_features:
        if col1 == col2:
            corr_matrix.loc[col1, col2] = 1.0
        else:
            corr_matrix.loc[col1, col2] = cramers_v(df[col1], df[col2])

corr_matrix = corr_matrix.astype(float)
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar=True)
plt.title('Correlation Matrix of Categorical Features (Cramer\'s V)')
plt.tight_layout()
plt.savefig('categorical_correlation_heatmap.png')
plt.show()
```

Cramer’s V values range from 0 (no association) to 1 (perfect association), with the heatmap using a color gradient (blue = low, red = high). The diagonal (e.g., HighBP vs. HighBP) is 1.0 (perfect self-correlation), and the matrix is symmetric. 


From the correlation matrix graph above, it is shown that none of the features are greatly correlated to each other. However, High blood pressure and difficulty walking strongly correlate with diabetes (0.35, 0.32). Age and income tie to diabetes and education (0.29, 0.22), reflecting disparities. Lifestyle factors like physical activity link to fruit and veggie intake (0.25), suggesting intervention areas.

```python
# Subsample 20% for hyperparameter tuning
X_small, _, y_small, _ = train_test_split(
    X_train_resampled, y_train_resampled, train_size = 0.2, stratify=y_train_resampled, random_state = RANDOM_STATE
)
```

```python
skf = StratifiedKFold(n_splits = 3, shuffle = True, random_state = RANDOM_STATE)

best_models = {}
```

```python
for name, (model, param_grid) in models_and_params.items():
    print(f"\n🔍 Tuning {name}...")
    try:
        grid = GridSearchCV(
            model,
            param_grid=param_grid,
            cv=skf,
            scoring='roc_auc',
            n_jobs=-1,
            verbose=1
        )
        grid.fit(X_small, y_small)

        print(f"✅ Best parameters for {name}: {grid.best_params_}")

        # Retrain with best parameters on full resampled data
        best_model = grid.best_estimator_
        best_model.fit(X_train_resampled, y_train_resampled)

        # Predict and evaluate
        y_pred = best_model.predict(X_test)
        y_proba = best_model.predict_proba(X_test)[:, 1]

        acc = accuracy_score(y_test, y_pred)
        auc = roc_auc_score(y_test, y_proba)

        print(f"{name} Test Accuracy: {acc:.4f}, ROC AUC: {auc:.4f}")

        best_models[name] = {
            'model': best_model,
            'accuracy': acc,
            'roc_auc': auc,
            'params': grid.best_params_
        }

    except Exception as e:
        print(f"❌ Error with {name}: {e}")
```

```python
# Loop through best models and evaluate on X_test
for name, info in best_models.items():
    print(f"\n🧪 Evaluating {name} on test set...")

    model = info['model']
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    # Confusion Matrix & Classification Report
    cm = confusion_matrix(y_test, y_pred)
    report = classification_report(y_test, y_pred, digits=4)
    roc_auc = roc_auc_score(y_test, y_proba)

    print("Confusion Matrix:")
    print(cm)
    print("\nClassification Report:")
    print(report)
    print(f"ROC AUC Score: {roc_auc:.4f}")

    # Plot Confusion Matrix
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=["No Diabetes", "Diabetes"], yticklabels=["No Diabetes", "Diabetes"])
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()

    # Plot ROC Curve
    fpr, tpr, thresholds = roc_curve(y_test, y_proba)
    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, label=f"{name} (AUC = {roc_auc:.4f})")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"{name} - ROC Curve")
    plt.legend(loc="lower right")
    plt.grid()
    plt.tight_layout()
    plt.show()
```
